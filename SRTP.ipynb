```python
def tfidfGenerating(text):
    wordpool=[]
    for word in newTfidfTags:
        if word[0] in text:
            wordpool.append(word)
    length=len(wordpool)
    index = random.sample(range(0,length),1)
    holeWord=wordpool[index[0]][0]
    wordtype=wordpool[index[0]][1]
    newString=""
    newString+=text[0:text.find(holeWord)]
    newString+="__________"
    newString+=text[text.find(holeWord)+len(holeWord):]
    generatedSelection=[]
    typeCollection=[]
    for tag in newTfidfTags:
        if tag[1]==wordtype and tag[0]!=holeWord:
            typeCollection.append(tag[0])
    index = random.sample(range(0,len(typeCollection)),3)
    for ind in index:
        generatedSelection.append(typeCollection[ind])
    generatedSelection.append(holeWord)
    random.shuffle(generatedSelection)
    print(newString)
    print(generatedSelection)
    print(holeWord)
```
